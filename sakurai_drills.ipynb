# Modern Quantum Mechanics — Sakurai drills in Python

Deze notebook volgt de kernideeën uit Sakurai's *Modern Quantum Mechanics* maar direct in Python/NumPy.

We gebruiken de module `quant_math_lab.py` die alle basiselementen (Dirac-notatie, Pauli-algebra, onzekerheidsrelatie, Bell-toestanden, Deutsch/DJ, QPE) implementeert.

```python
import numpy as np
from quant_math_lab import *
```

## 1. Bras, kets en projectoren

```python
# |psi> = (1, i)/√2
psi = ket([1,1j])
print("psi =\n", psi)

# Projector |psi><psi|
P = projector(psi)
print("Projector:\n", P)
print("Hermitisch:", np.allclose(P, P.conj().T))
```

## 2. Pauli-algebra

```python
from quant_math_lab import X,Y,Z
print("[X,Y] =\n", commutator(X,Y))
print("2iZ =\n", 2j*Z)
```

## 3. Onzekerheidsrelatie

```python
plus = ket([1,1])
dX, dZ, bound = uncertainty_bound(X,Z,plus)
print("ΔX=",dX," ΔZ=",dZ," bound=",bound)
```

## 4. Bell-staat en partiële spoor

```python
phi = (tensor(ket([1,0]), ket([1,0])) + tensor(ket([0,1]), ket([0,1])))/np.sqrt(2)
rho = dm(phi)
red = partial_trace(rho, dims=[2,2], traced_out=[1])
print("Gerduceerde toestand A =\n", red)
```

## 5. CHSH-waarde

```python
print("CHSH(|phi+>) =", chsh_value(rho))
```

## 6. Deutsch en Deutsch–Jozsa

```python
Uf = deutsch_oracle("identity")
print("Deutsch orakel identity ->", deutsch_algorithm(Uf))

Zf = dj_oracle(4, "balanced")
print("Deutsch–Jozsa (balanced) ->", deutsch_jozsa(Zf))
```

## 7. Quantum Phase Estimation (simulatie)

```python
theta = 11/32
U = np.diag([1, np.exp(2j*np.pi*theta)])
y, approx = qpe_simulate(U, ket([0,1]), m=6)
print("QPE resultaat:", y, "/",2**6, "≈", approx, "(true", theta, ")")
```
