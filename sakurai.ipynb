The time-reversal operator $\Theta$ is an antiunitary operator. 
It can be written as the product of a unitary rotation and complex conjugation. 
In Sakurai’s {Modern Quantum Mechanics} (Ch.~4), this is given (up to a phase) by:
$$
\Theta = e^{-i\pi J_y/\hbar} K,
$$
where $J_y$ is the $y$-component of the  {total angular momentum} (including spin), 
and $K$ is the complex-conjugation operator. 
The $180^\circ$ rotation about the $y$-axis ensures that  {all components} of angular momentum flip sign under $\Theta$, 
while $K$ handles the reversal of momenta and complex phases. 

For a spinless particle, $\Theta$ reduces to just $K$, since complex conjugation $\psi(\mathbf{r}) \to \psi^*(\mathbf{r})$ 
already sends $\mathbf{p} \to -\mathbf{p}$. For a particle with spin $s$, one finds that $\Theta$ sends each spin basis ket to the one 
with  {opposite} $m$: 
$$
\Theta |{s,m}\rangle = c_m |{s,-m}\rangle,
$$
for some phase $c_m$. A specific example is spin-$\tfrac{1}{2}$, where $e^{-i\pi S_y/\hbar} = -i\sigma_y$ flips $\ket{\uparrow} \leftrightarrow \ket{\downarrow}$, 
and $K$ then yields the usual $\Theta |{\uparrow}\rangle \propto |{\downarrow}\rangle$, $\Theta |{\downarrow}\rangle \propto |{\uparrow}\rangle$.

Factorizing into Orbital and Spin Parts

It is valid to separate the orbital and spin parts when writing $\Theta$ for a system with both  {orbital} angular momentum $L$ and  {spin} $S$. 
In a product basis $\ket{l,m_l} \otimes \ket{s,m_s}$, the total $J_y$ acts as
$$
J_y = L_y \otimes I + I \otimes S_y.
$$

Crucially, $L_y$ and $S_y$ commute because they act on independent subspaces (orbital vs. spin). 
When two operators commute, the exponential of their sum factorizes exactly:
$$
e^{-i\pi (L_y + S_y)/\hbar} = e^{-i\pi L_y/\hbar} \, e^{-i\pi S_y/\hbar}.
$$

Therefore, we can write the time-reversal operator as:
$$
\Theta = e^{-i\pi L_y/\hbar} \, e^{-i\pi S_y/\hbar} \, K,
$$
(up to an overall phase). 
Here $e^{-i\pi L_y/\hbar}$ acts on the orbital wavefunction, and $e^{-i\pi S_y/\hbar}$ acts on the spin state. 
This factorized form is completely  {equivalent} to Sakurai’s $e^{-i\pi J_y/\hbar}K$ form, 
since $J_y = L_y + S_y$. It simply makes explicit that the $180^\circ$ rotation is composed of an orbital rotation 
and a spin rotation, which  {commute} and can be applied independently. 

Implications for L.S and Central Potenials

This separation is not only valid but conceptually helpful in typical cases like a  {central potential with spin} 
or a  {spin–orbital coupling} Hamiltonian. For example, consider a Hamiltonian with a spin–orbit term $H_{\text{SO}} \propto \mathbf{L}\cdot\mathbf{S}$. 
Under time reversal,  {$\mathbf{L}$ reverses sign and $\mathbf{S}$ reverses sign}, so their dot product is invariant:
$$
\Theta \, (\mathbf{L}\cdot\mathbf{S}) \, \Theta^{-1} = (-\mathbf{L})\cdot(-\mathbf{S}) = \mathbf{L}\cdot\mathbf{S}.
$$

This means $H_{\text{SO}}$ commutes with $\Theta$ (time-reversal symmetry is preserved). 
Indeed, any central potential $V(r)$ with a kinetic term $\mathbf{p}^2/2m$ is time-reversal invariant, 
and including a spin degree of freedom does not break T-symmetry as long as no external magnetic field 
or T-odd terms are present.

In the basis $\ket{l,m_l; s,m_s}$, applying $\Theta$ as $e^{-i\pi L_y/\hbar} \, e^{-i\pi S_y/\hbar} \, K$ 
will send $\ket{l,m_l; s,m_s}$ to (a phase times) $\ket{l,-m_l; s,-m_s}$, 
which is consistent with how one expects time reversal to act.
